```python
cur=conn.cursor() #connecting to the database for the operations,
#Cursors represent a database cursor (and map to ODBC HSTMTs), 
#which is used to manage the context of a fetch operation. 
#Cursors created from the same connection are not isolated, i.e., 
#any changes done to the database by a cursor are immediately visible by the other cursors.

```

```python
cur.execute('''CREATE TABLE COMPANY
       (ID INT PRIMARY KEY     NOT NULL,
       NAME           TEXT    NOT NULL,
       AGE            INT     NOT NULL,
       ADDRESS        CHAR(50),
       SALARY         REAL);''')
conn.commit()
conn.close()
# creates the tables in the local postgres spatial database #executes the sql query
```
